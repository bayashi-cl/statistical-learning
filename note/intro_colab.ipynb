ノートブックはセルの集まりからなっています。セルには説明用のテキストか、もしくは実行可能なコードとその出力が含まれます。クリックすることでそのセルを選択できます。


## コードセル
この下のセルが**コードセル**です。ツールバー右上にあるボタンが接続の状態を表していて、接続されていればセルをクリックして選択し、以下の方法で実行することができます。

* セルの左端にある**再生ボタン**をクリックする。
* **Cmd/Ctrl+Enter** で選択されたセルが実行される。
* **Shift+Enter** でセルが実行され、フォーカスが次のセルに移る。（もし後ろにセルが無ければ新たなセルが追加される。）
* **Alt+Enter** でセルが実行され、新たなコードセルが下に追加される。

**ランタイム**メニューから複数のセルをまとめて実行する事ができます。

## テキストセル
このセルが**テキストセル**です。**ダブルクリック**をするとセルの内容を編集することができます。テキストセルではmarkdown記法が使われています。詳しくは[markdown

## セルの追加と削除
セルの間にマウスカーソルを置くと表示される**＋コード** **＋テキスト**ボタンを使って、新しいセルを追加できます。これらのボタンは、ノートブックの上のツールバーにもあり、現在選択されているセルの下にセルを追加できます。

セルの移動は、セル右上のツールバーの「セルを上に移動」または「セルを下に移動」をクリックして行います。

連続したセルを選択するには、セルの外側からグループ内をドラッグする「投げ縄選択」を使用します。隣り合っていないセルを同時に選択するには、あるセルをクリックした後、Ctrlを押しながら別のセルをクリックします。同様に、Ctrlの代わりにShiftを使用すると、間のセルがすべて選択されます。

## リッチでインタラクティブな出力
これまで生成された出力はすべてテキストでしたが、以下ののように、より興味深いものにすることができます。

```r
library(ggplot2)
data <- data.frame(
  ys <- 200 + rnorm(100),
  x <- 0:(length(ys) - 1)
) 
g <- ggplot(data, aes(x = x, y = ys)) + ggtitle("Fills and Alpha Example") + theme(plot.title = element_text(hjust = 0.5))
g <- g + geom_line(color = "royalblue") + geom_area(fill = "darkgreen", alpha=0.6)
g <- g + coord_cartesian(ylim = c(195, 203))
plot(g)
```

## googleドライブとの連携
ColaboratoryはGoogle Driveと連携しています。これにより、同じドキュメントを複数の人と共有したり、コメントしたり、共同作業を行うことができます。

シェアボタン（ツールバーの右上）では、ノートブックを共有したり、ノートブックに設定された権限を制御することができます。

「ファイル」->「ドライブにコピーを保存」で、ノートブックのコピーをドライブに作成します。

「ファイル」->「保存」で、Driveにファイルを保存します。

「ファイル」->「変更履歴」で、ノートブックの変更履歴を表示します。

## セルにコメントする
Colaboratory ノートブックには、Google Document と同様にコメントを付けることができます。コメントはセルに添付され、参照しているセルの横に表示されます。コメントのみの権限を持っている場合は、セルの上にマウスを置くと右上にコメントボタンが表示されます。

編集権限またはコメント権限を持っている場合は、3つの方法のいずれかでセルにコメントを付けることができます。

1. セルを選択して、セルの右上にあるツールバーのコメントボタンをクリックする。
1. テキストセルを右クリックし、コンテキストメニューから「コメントを追加」を選択する。
1. ショートカット（Ctrl+Shift+M）を使って、現在選択されているセルにコメントを追加する。

コメントを解決したり、コメントに返信したりすることができます。また、+[メールアドレス] (例: +user@domain.com) と入力することで、コメントを特定の共同作業者に向けて送信することができます。アドレスを指定した共同研究者にはメールが送信されます。

ページの右上にある「コメント」ボタンには、ノートブックに付けられたすべてのコメントが表示されます
